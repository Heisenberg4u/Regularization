```python
LabourData_num = LabourData[['Age','Nodeg', 'Earnings_1974', 'Earnings_1975', 'Earnings_1978']]
```

```python
LabourData_dummies = pd.get_dummies(LabourData[['Race', 'Hisp', 'MaritalStatus', 'Eduacation']])
LabourData_dummies.head()
```

```python
LabourData_combined = pd.concat([LabourData_num, LabourData_dummies], axis=1)
LabourData_combined.head()
```

```python
# Putting feature variable to X
X = LabourData_combined[['Age', 'Earnings_1974', 'Earnings_1975', 'Race_NotBlack', 'Race_black', 
                         'Hisp_NotHispanic', 'Hisp_hispanic','MaritalStatus_Married', 
                         'MaritalStatus_NotMarried', 'Eduacation_HighSchool', 'Eduacation_Intermediate',
                         'Eduacation_LessThanHighSchool', 'Eduacation_PostGraduate', 'Eduacation_graduate']]

# Putting response variable to y
y = LabourData['Earnings_1978']
```

```python
#random_state is the seed used by the random number generator. It can be any integer.
from sklearn.model_selection import train_test_split, GridSearchCV
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7 , random_state=100)
```

```python
# Define Model parameters to tune
model_parameters = {
                    'alpha':[1, 10, 100, 1000],
                    'normalize': [False, True] # Standardscaler, variable-mean/std.dev
                    }
```

```python
# Gridsearch the parameters to find the best parameters.
from sklearn.linear_model import Lasso

model = Lasso(random_state=1)

gscv = GridSearchCV(estimator=model, 
                    param_grid=model_parameters, 
                    cv=5,          # 5-Fold Cross Validation
                    verbose=1,      #To print what it is doing
                    n_jobs=-1,      #fastest possible depending in the laptop
                    scoring='r2')     #r2 value for regression, higher the better
gscv.fit(X_train, y_train)
```

```python
print('The best parameter are -', gscv.best_params_)
```

```python
print(gscv.best_score_)
print(gscv.best_estimator_)
print(gscv.scorer_)
```

```python
# Re-fit the model with the best parameters
final_mod = Lasso(**gscv.best_params_, random_state=1)
final_mod.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
print('R square on test by final_mod =', r2_score(y_true=y_test,
                                                        y_pred=final_mod.predict(X_test)))
```

```python
print('R square on test by gscv =', r2_score(y_true=y_test,
                                                        y_pred=gscv.predict(X_test)))
```

```python
# View the model coefficients
list(zip(X_train.columns, final_mod.coef_))
```

```python
print(final_mod.predict(X_test)[:5])
print(y_test.to_numpy()[:5])
```
